```python
import matplotlib.pyplot as plt
import numpy as np
```

Tạo ra 1 vector x có 1500 phần tử với giá trị khởi đầu là -6 và kết thúc là 5

```python
x = np.linspace(-6, 5, num = 1500)
```

Xây dựng vector y, y1 và y2 tương ứng hàm số y(x), y(x)
′
, y(x)
′′ với giá trị các phần tử

tương ứng được sinh ra từ việc sử dụng vector x ở trên làm tham số đầu vào;

```python
y = x**3 + 3*x + 7
y1 = 3*x**2 + 3
y2 = 6*x
```

- Hãy tạo đối tượng fig (là đối tượng của lớp Figure) và đối tượng ax (là đối tượng của
Axes) từ hàm subplot() để phục vụ việc vẽ hình đồ thị của các hàm số trên;
- Sử dụng đối tượng ax để vẽ đồ thị cho hàm số y(x) tương ứng với tham số đầu vào là x và
y và đặt tên cho đồ thị này là y(x);

- Sử dụng đối tượng ax để vẽ đồ thị cho hàm số y
′
(x) tương ứng với tham số đầu vào là x và y1 và đặt tên cho đồ thị này là y
′
(x);
- Sử dụng đối tượng ax để vẽ đồ thị cho hàm số y′′(x) tương ứng với tham số đầu vào là x và y2 và đặt tên cho đồ thị này là y′′(x);

- Thiết lập tên cho các trục tọa độ: tung độ là y và hoành độ là x;
- Hiển thị các tên của từng đồ thị lên hình vẽ

```python
fig, ax = plt.subplots()
ax.plot(x, y, label="y(x)")
ax.set_title("y")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y1, label="y(x)'")
ax.set_title("y(x)'")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y2, label='y(x)"')
ax.set_title('y(x)"')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
```

Kết quả xuất ra file ảnh định dạng svg

```python
fig, ax = plt.subplots()
ax.plot(x, y, label="y(x)")
ax.plot(x, y1, label="y(x)'")
ax.plot(x, y2, label='y(x)"')
fig.savefig('myimage.svg', format='svg', dpi=1200)
```
